```python
import cv2
import easyocr
import matplotlib.pyplot as plt
import re
```

```python
# Read image
image_path = "/content/eg.png"
img = cv2.imread(image_path)
```

```python
# Instantiate text detector
reader = easyocr.Reader(["en"], gpu=False)
```

```python
# Detect text on image
text_ = reader.readtext(img)

threshold = 0.25

# Define a regular expression pattern to match alphabetic characters
alpha_pattern = re.compile(r'[A-Za-z1-9]+')-
```

```python
# Draw bounding boxes and text, and print the recognized text (alphabetic only)
for t_, t in enumerate(text_):
    if t_ == 0:
        continue  # Skip the first item which corresponds to the index
    bbox, text, score = t
    if score > threshold:
        # Check if the text contains alphabetic characters
        if alpha_pattern.match(text):
            # Extract the top-left and bottom-right coordinates of the bounding box
            pt1 = tuple(map(int, bbox[0]))
            pt2 = tuple(map(int, bbox[2]))

            # Draw bounding box and text on the image
            cv2.rectangle(img, pt1, pt2, (0, 255, 0), 5)
            cv2.putText(img, text, pt1, cv2.FONT_HERSHEY_COMPLEX, 2, (255, 0, 0), 2)

            # Print recognized text to the terminal
            print(text)
```

```python
# Display the image with bounding boxes and text
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```
