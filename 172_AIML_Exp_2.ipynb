```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
columns_to_visualize = ['Age', 'Gender', 'Total_Bilirubin', 'Direct_Bilirubin',
                        'Alkaline_Phosphotase', 'Alamine_Aminotransferase',
                        'Aspartate_Aminotransferase', 'Total_Protiens', 'Albumin',
                        'Albumin_and_Globulin_Ratio', 'Dataset']

# Set the style of seaborn plots
sns.set(style='whitegrid')

# Create subplots
fig, axes = plt.subplots(nrows=4, ncols=3, figsize=(15, 15))
fig.subplots_adjust(hspace=0.5)

# Iterate over selected columns and create visualizations
for col, ax in zip(columns_to_visualize, axes.flatten()):
    if col == 'Gender':
        sns.countplot(x=col, data=df, ax=ax)
    else:
        sns.histplot(df[col], ax=ax, kde=True)
        ax.set_xlabel(col)

plt.tight_layout()
plt.show()
```

```python
columns_to_include = ['Age', 'Total_Bilirubin', 'Direct_Bilirubin',
                      'Alkaline_Phosphotase', 'Alamine_Aminotransferase',
                      'Aspartate_Aminotransferase', 'Total_Protiens', 'Albumin',
                      'Albumin_and_Globulin_Ratio']

# Compute the correlation matrix
correlation_matrix = df[columns_to_include].corr()

# Set up the matplotlib figure
plt.figure(figsize=(10, 8))

# Create a heatmap using seaborn
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)

plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(6, 4))

# Create a bar plot for the 'Age' column
sns.countplot(x='Age', data=df)

plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Bar Graph of Age Distribution')
plt.tight_layout()
plt.show()
```

Violin plot for the 'Albumin' column

```python
sns.violinplot(y=df.Albumin)
```

```python
gender_counts = df['Gender'].value_counts()

# Set up the matplotlib figure
plt.figure(figsize=(6, 6))

# Create a pie chart for the 'Gender' column
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=140)

plt.title('Pie Chart of Gender Distribution')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

```python
plt.figure(figsize=(10, 6))

# Create a line plot for 'Age' vs 'Total_Bilirubin'
sns.lineplot(x='Age', y='Total_Bilirubin', data=df)

plt.xlabel('Age')
plt.ylabel('Total Bilirubin')
plt.title('Line Plot: Age vs Total Bilirubin')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))

# Create a line plot for 'Age' vs 'Total_Bilirubin'
sns.lineplot(x='Age', y='Total_Bilirubin', data=df)

plt.xlabel('Age')
plt.ylabel('Albumin')
plt.title('Line Plot: Age vs Albumin')
plt.tight_layout()
plt.show()
```
