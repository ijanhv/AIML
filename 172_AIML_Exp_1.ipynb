```python
dataframe1 = pd.DataFrame(df)
dataframe2 = pd.DataFrame(df1)

# Perform the join operation based on the common attribute 'ID'
merged_df = dataframe1.merge(dataframe2, on='Age', how='inner')
merged_df.head()
```

```python
# Removing noise from dataset having column 'Dataset' which has no relevance with the other attributes
merged_df = merged_df.drop('Dataset', axis=1)
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
merged_df.columns
```

```python
X = merged_df.drop('Outcome', axis=1)
y = merged_df['Outcome']

# Create a MinMaxScaler instance
scaler = MinMaxScaler()

# Normalize the 'Outcome' column
y_normalized = scaler.fit_transform(y.values.reshape(-1, 1))

# Convert the normalized values back to a DataFrame
normalized_y = pd.DataFrame(y_normalized, columns=['Outcome'])

# Concatenate normalized 'Outcome' column with features
normalized_data = pd.concat([X, normalized_y], axis=1)
```
